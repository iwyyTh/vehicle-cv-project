```python
import numpy as np
import matplotlib.pyplot as plt

print("NumPy version: ", np.__version__)
```

```python
anh_gia = np.random.randint(0,256, size = (100,100,3), dtype= np.uint8)

print("Kieu du lieu: ", anh_gia.dtype)
print("Shape: ", anh_gia.shape)
print("H, W, C: ", anh_gia.shape[0], anh_gia.shape[1], anh_gia.shape[2])
print("Tong so pixel: ", anh_gia.shape[0]* anh_gia.shape[1])
```

```python

plt.figure(figsize=(4,4))
plt.imshow(anh_gia)
plt.title("Anh ngau nhien 100x100")
plt.axis("off")
plt.show()
```

```python
goc_tren_trai = anh_gia[0:30, 0:30, :]
kenh_do = anh_gia[:,:,0]
hang_dau_tien = anh_gia[0,:,:]

print("Vung ROI shape: ", goc_tren_trai.shape)
print("Kênh đỏ shape  :", kenh_do.shape)         # (100, 100)
print("Hàng đầu shape :", hang_dau_tien.shape) 
```

```python
anh_float = anh_gia.astype(np.float32)/255.0

print("Truoc normalize: ", anh_gia.min(), "->", anh_gia.max(), anh_gia.dtype)
print("Sau normalize  :", anh_float.min(), "→", anh_float.max(), anh_float.dtype)
```

```python
fig, axes = plt.subplots(1,3,figsize=(12,3))
mau_sac = ['Đỏ (R)', 'Lục (G)', 'Lam (B)']
colors  = ['red', 'green', 'blue']
for i in range(3):
    axes[i].hist(anh_gia[:,:,i].ravel(), bins=50, color=colors[i], alpha=0.7)
    axes[i].set_title(mau_sac[i])
    axes[i].set_xlabel("Gia tri pixel (0-255)")
    axes[i].set_ylabel("So luong pixel")
plt.tight_layout()
plt.show()
```
